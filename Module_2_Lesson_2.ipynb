## STABILITY AND THE CFL CONDITION

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python

pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
def linear_convection(nx, L=2.0, c=1.0, dt=0.025, nt=20):
    
    dx = L / (nx - 1)
    x = numpy.linspace(0.0, L, num=nx)
    
    u0 = numpy.ones(nx)
    mask = numpy.where((x >= 0.5) & (x <= 1.0))
    u0[mask] = 2.0
    
    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])
    
    pyplot.figure(figsize=(10.0, 5.0))
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial', color='C0', linestyle='-', linewidth=2)
    pyplot.plot(x, u, label='Current', color='C1', linestyle='-', linewidth=2)
    pyplot.legend()
    pyplot.xlim(0.0, L)
    pyplot.ylim(0.0, 2.5);
```

```python
linear_convection(41)  # solve using 41 spatial grid points
```

```python
linear_convection(61)
```

```python
linear_convection(71)  # solve using 41 spatial grid points
```

```python
linear_convection(85)
```

```python
def linear_convection_cfl(nx, L=2.0, c=1.0, sigma=0.5, nt=20):
    
    dx = L / (nx - 1)
    x = numpy.linspace(0.0, L, num=nx)
  
    dt = sigma * dx / c

    u0 = numpy.ones(nx)
    mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0

    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])

    pyplot.figure(figsize=(10.0, 5.0))
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial',
                color='C0', linestyle='--', linewidth=2)
    pyplot.plot(x, u, label='nt = {}'.format(nt),
                color='C1', linestyle='-', linewidth=2)
    pyplot.legend()
    pyplot.xlim(0.0, L)
    pyplot.ylim(0.0, 2.5);
```

```python
linear_convection_cfl(85)
```

```python
linear_convection_cfl(141)
```
